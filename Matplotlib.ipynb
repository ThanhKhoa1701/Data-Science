### Pyplot API vs Object-Oriented(OO) API
* pylot - Quickly
* OO API - Advanced 

```python
# pyplot API
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = 'blue', linestyle = 'dashed', label = 'sin(x)')
plt.plot(x, np.cos(x), color = 'red', label ='cos(x)')

plt.title('Sin(x) and Cos(X)')
plt.xlabel('X')
plt.ylabel('F(X)')

#plt.xlim([0,4])
#plt.axis([xmin, xmax, ymin, ymax])
#plt.axis('tight')
plt.axis('equal')

plt.legend();
```

```python
# object oriented OO API
# 1. Prepare data
x= [2,3,4,5]
y = [45,3,67,22]

#2. Setup plot
fig, ax = plt.subplots(figsize= (5,5)) #Figsize = width & height of the plot

#3. Plot the data
ax.plot(x,y)

ax.set( title = 'A simple plot', xlabel = 'x-axis', ylabel= 'y-axis');
```

### Histogram

### Subplots

```python
# Option1:
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows= 2, ncols = 2, figsize = (10,5))
#plot data
ax1.plot(x,x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.rand(10), np.random.rand(10), color = 'green');

ax3.bar(Drink_prices.keys(), Drink_prices.values(), color = 'blue' );

ax4.hist(student_height, bins = 30, color = 'red');
```

```python
## Option2:
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize= (10,5))

ax[0,0].plot(x,x/2);
ax[0,0].plot(x, x*2);

ax[0,1].scatter(np.random.rand(10), np.random.rand(10), color = 'green');

ax[1,0].bar(Drink_prices.keys(), Drink_prices.values(), color = 'blue' );

ax[1,1].hist(student_height, bins = 30, color = 'red');
```

### Data Visualization - California Cities

```python
cities = pd.read_csv('california_cities.csv')
cities.head()
```

```python
lat, lon , population, area = cities['latd'], cities['longd'], cities['population_total'], cities['area_total_km2']

plt.style.use('seaborn-v0_8')
plt.figure(figsize = (10,8))

plt.scatter(lon,lat, c=np.log10(population), cmap= 'viridis',
           s = area, linewidths = 0, alpha =0.7)

plt.axis('equal')
plt.ylabel('latitude')
plt.xlabel('longlatitude')
plt.colorbar ( label ='log$_{10}$(population)')
plt.clim(3,7)

# create a lengend 
area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([], [], s= area, color ='k', alpha = 0.4,
               label = str(area) + 'km$^2$')
    plt.legend(labelspacing =1 , title = 'City Area')
plt.title('California cities population and area distribution');
```
